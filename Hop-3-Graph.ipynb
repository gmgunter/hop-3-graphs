```python
import networkx as nx
import matplotlib.pyplot as plt

# Number of SAR images
N = 10

# Number of scenes it is connected to in the future
bw = 3

g = nx.Graph()
for ii in range(N):
    g.add_node(ii)

for ii in range(N):
    for jj in range(ii + 1, min(N, ii + bw + 1)):
        g.add_edge(ii, jj)

plt.figure()
nx.draw_planar(g, with_labels=True)
plt.show()
```
